# Metric Definition Function

```python
def wmae_test(test, pred): # WMAE for test
    weights = X_test['IsHoliday'].apply(lambda is_holiday:5 if is_holiday else 1)
    error = np.sum(weights * np.abs(test - pred), axis=0) / np.sum(weights)
    return error
```

###**Penjelasan kode di atas**

Fungsi wmae_test(test, pred) digunakan untuk menghitung Weighted Mean Absolute Error (WMAE), yaitu metrik evaluasi yang memberikan bobot lebih besar pada kesalahan prediksi saat hari libur. Dalam fungsi ini, setiap nilai pada kolom IsHoliday di X_test dikonversi menjadi bobot, yaitu 5 untuk hari libur dan 1 untuk hari biasa. Kemudian, kesalahan absolut antara nilai aktual dan prediksi dikalikan dengan bobot tersebut, dijumlahkan, dan dibagi dengan total bobot untuk memperoleh rata-rata kesalahan berbobot.
